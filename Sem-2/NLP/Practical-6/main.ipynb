```python
import spacy
# Load the pre-trained spaCy model for English
nlp = spacy.load("en_core_web_sm")
# Function to perform dependency parsing
def dependency_parsing(text):
    # Process the text using spaCy
    doc = nlp(text)
    # Print the word, its part-of-speech, and its dependency relation
    print(f"{'Word':<15} {'POS':<10} {'Dependency Relation'}")
    print("="*40)
    for token in doc:
        print(f"{token.text:<15} {token.pos_:<10} {token.dep_:<20}")
    # Visualize the dependency tree (optional)
    # If you want to display the dependency tree in a Jupyter notebook or graphical environment:
    # doc.visualize()

# Example text to parse
text = "The quick brown fox jumps over the lazy dog."
# Perform dependency parsing
dependency_parsing(text)
```
